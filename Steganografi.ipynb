```python
from PIL import Image
from stegano import lsb
import io
from google.colab import files

def hide_message(image_path, message):
    """ Menyisipkan pesan ke dalam gambar """
    image = Image.open(image_path)
    secret_image = lsb.hide(image, message)

    # Simpan hasil gambar
    output_path = "secret_image.png"
    secret_image.save(output_path)

    print("‚úÖ Pesan telah disisipkan dalam gambar.")
    return output_path

def reveal_message(image_path):
    """ Mengambil pesan tersembunyi dari gambar """
    extracted_message = lsb.reveal(image_path)
    return extracted_message if extracted_message else "Tidak ada pesan tersembunyi ditemukan."

# Pilih mode (1: Menyisipkan pesan, 2: Mengambil pesan)
mode = input("Pilih mode (1: Sisipkan Pesan, 2: Ambil Pesan): ")

if mode == "1":
    # Menyisipkan pesan
    image_path = input("Masukkan nama file gambar yang telah di-upload: ")
    message = input("Masukkan pesan rahasia: ")
    secret_image_path = hide_message(image_path, message)

    # Unduh gambar dengan pesan tersembunyi
    files.download(secret_image_path)

elif mode == "2":
    # Mengambil pesan
    image_path = input("Masukkan nama file gambar yang telah di-upload: ")
    extracted_message = reveal_message(image_path)
    print(f"üì© Pesan tersembunyi: {extracted_message}")

else:
    print("‚ùå Pilihan tidak valid. Silakan jalankan ulang program.")
```
